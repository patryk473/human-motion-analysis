## 4. Matematyka

```python
#Odległość między punktami
def distance(P1,P2):
    distance = np.linalg.norm(P1 - P2)

#Kąt z trzech punktów
def angle_3_points(A, B, C):
    BA = A - B
    BC = C - B

    cosine_angle = np.dot(BA, BC) / (
        np.linalg.norm(BA) * np.linalg.norm(BC)
    )

    angle = np.degrees(np.arccos(cosine_angle))
    return angle
```
